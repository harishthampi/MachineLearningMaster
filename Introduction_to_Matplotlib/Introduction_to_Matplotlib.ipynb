## Matplotlib WorkFlow Example

```python
#1.Import matplotlib 
import matplotlib.pyplot as plt
#2.Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]
#3.Setup plot (Figure and Axes)
fig, ax = plt.subplots(figsize=(10,10)) #width,height of the figure
#4.Plot data
ax.plot(x,y)
#5.Customization
ax.set(title = "Sample Data",xlabel="x-axis",ylabel="y-axis")
#6.Save the image
fig.savefig("images/simple-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title = "NutButter Prices", ylabel= "Prices ($)");
```

```python
# horizontal bar-graph
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
# Create Data and make a Histogram graph
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for multiple Axes with subplots

```python
# Subplot option1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,ncols = 2,figsize=(10,5))
#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.randn(100),np.random.randn(100));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(100))
```

```python
#Subplots option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randn(100),np.random.randn(100));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100));
```

## Plotting from pandas DataFrame

```python
# Make a DataFrame
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Convert the Price to an integer and remove $,. 
car_sales["Price"] = car_sales["Price"].str.replace(r'[$,.]', '', regex=True)
#removes the $ , and . characters from every value in the Price column of your DataFrame
car_sales
```

```python
# Remove the extra zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#add a date column to dataframe
car_sales["Sales Date"] = pd.date_range("1/1/2025",periods = len(car_sales))
car_sales
```

```python
# add a new columnas Total sales 
car_sales["Price"] = car_sales["Price"].astype(int)
# convert the price field from string to int
car_sales["Total Sales"] = car_sales["Price"].cumsum()
# cumsum() -> calculates the cumulative sum of elements in a sequence, array, or series
car_sales
```

```python
# Plot Total Sales vs Sales Date
car_sales.plot(x="Sales Date",y="Total Sales");
```

```python
# Plot Odometer (KM) vs Price
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
#import another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist();
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,40),subplots=True);
```

```python
# Plot age vs chol based on target for over the age of 50

over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50.head()
```

```python
# pyplot method
over_50.plot.scatter(x="age",y="chol",c="target",figsize=(10,6))
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#Customize the plot
ax.set(title = "Heart Disease and Cholestrol Levels",xlabel = "Age",ylabel = "Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target") #scatter.legend_elements() automatically generates handles and labels for the unique values in the c argument you passed to scatter.
#handles, labels = scatter.legend_elements()
#ax.legend(handles, labels, title="Target")

#add horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle='--')
```

```python
# Create a subplot with age,chol,thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)
#Prepare data on ax0,

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

#Customize plot
ax0.set(title = 'Heart Disease and Cholestrol Levels',
ylabel='Cholestrol')

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target")

#Add mean Line
ax0.axhline(y=over_50["chol"].mean(),linestyle='--')

#Prepare data on ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

#Customize plot
ax1.set(title = 'Heart Disease and Thalach Levels',
ylabel='Thalach')

#Add legend
ax1.legend(*scatter.legend_elements(),title="Target")

#Add mean Line
ax1.axhline(y=over_50["thalach"].mean(),linestyle='--')

#Add figure Title
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight=("bold"));
```

```python
# Change the style of the plot inside a given style- example - seaborn-v0_8-whitegrid
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter") #cmap changes the color scheme
#Customize the plot
ax.set(title = "Heart Disease and Cholestrol Levels",xlabel = "Age",ylabel = "Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target") #scatter.legend_elements() automatically generates handles and labels for the unique values in the c argument you passed to scatter.
#handles, labels = scatter.legend_elements()
#ax.legend(handles, labels, title="Target")

#add horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle='--')
```

```python
# Create a subplot with age,chol,thalach
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)
#Prepare data on ax0,

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");

#Customize plot
ax0.set(title = 'Heart Disease and Cholestrol Levels',
ylabel='Cholestrol')

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target")
#Change the x axis limits
ax0.set_xlim([50,80])
#Add mean Line
ax0.axhline(y=over_50["chol"].mean(),linestyle='--')

#Prepare data on ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                      cmap="winter");

#Customize plot
ax1.set(title = 'Heart Disease and Thalach Levels',
ylabel='Thalach')

#Add legend
ax1.legend(*scatter.legend_elements(),title="Target")

#Add mean Line
ax1.axhline(y=over_50["thalach"].mean(),linestyle='--')
#Change the x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#Add figure Title
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight=("bold"));
```
